Подключаем необходимые библиотеки

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Возьмем равномерное распределение на отрезке [2, 5] и сгенерируем из него выборку объема 1000

```python
a = 2
b = 5
uniform_rv = sts.uniform(a, b-a)
sample = uniform_rv.rvs(1000)
```

Построим гистограмму выборки и изобразим поверх неё теоретическую плотность нормального распределения на отрезке [2, 5]

```python
plt.hist(sample, density=True)
x = np.linspace(0,7,100)
pdf = uniform_rv.pdf(x)
plt.plot(x, pdf)
plt.title('сравнение гистограммы выборки с теоретической плотностью равномерного распределения')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Для нескольких n сгенерируем 1000 выборок объёма n и построим гистограммы распределений их выборочных средних. По центральной предельной теореме полученные распределения должны быть нормальными с параметрами, задаваемыми исходным равномерным распределением.

```python
for n in (5, 10, 50, 100):
    averages = [np.mean(uniform_rv.rvs(n)) for i in range(1000)]
    plt.title(f"Распределение выборочных средних на выборке объёма {n}")
    plt.xlabel("Значение выборочного среднего")
    plt.ylabel("Частота")
    plt.hist(averages, density=True)
        
    u = (a + b) / 2
    sigma = math.sqrt((b - a) ** 2 / (12 * n))
    norm_rv = sts.norm(u, sigma)
    x = np.linspace(0, 7, 100)
    plt.plot(x, norm_rv.pdf(x))
    plt.show()
```

Как видно из полученных результатов, эксперимент подтверждает центральную предельную теорему. Также стоит обратить внимание, что при увеличении объёма выборок распределение выборочных средних лучше описывается нормальным распределением.
